```python
import pickle
import sqlite3
import pandas as pd
```

```python
with open('df_institute', 'rb') as f:
    df_institute = pickle.load(f)
    
with open('df_term', 'rb') as f:
    df_term = pickle.load(f)
    
with open('df_school', 'rb') as f:
    df_school = pickle.load(f)
    
with open('df_class', 'rb') as f:
    df_class = pickle.load(f)
    
df_class = df_class.rename(columns={
    'institute_id': 'fk_institute_id',
    'term_id': 'fk_term_id',
    'class_id': 'class_id',
    'school_id': 'fk_school'})

# display(df_institute)
# display(df_term)
# display(df_school)
# display(df_class)
```

```python
! rm uh_enrollment.db
con = sqlite3.connect('uh_enrollment.db')

con.execute('PRAGMA foreign_keys = ON;');

con.execute('''
    CREATE TABLE institute (
    institute_id TEXT  NOT NULL PRIMARY KEY,
    institute TEXT NOT NULL);''')

con.execute('''
    CREATE TABLE term (
    term_id INTEGER NOT NULL PRIMARY KEY,
    academic INTEGER NOT NULL, 
    year INTEGER NOT NULL,
    semester TEXT NOT NULL, 
    extension BOOL NOT NULL,
    start TEXT NOT NULL);''')

con.execute('''
    CREATE TABLE school (
    school_id TEXT NOT NULL PRIMARY KEY,
    school TEXT NOT NULL);''')

con.execute('''
    CREATE TABLE class (
    fk_institute_id TEXT NOT NULL,
    fk_term_id INTEGER NOT NULL,
    class_id INTEGER NOT NULL,
    fk_school TEXT NOT NULL,
    course TEXT NOT NULL,
    title TEXT NOT NULL,
    enrolled INTEGER NOT NULL,
    updated TEXT NOT NULL,
    PRIMARY KEY (
        fk_institute_id,
        fk_term_id,
        class_id),
    FOREIGN KEY (fk_institute_id) REFERENCES institute(institute_id),
    FOREIGN KEY (fk_term_id) REFERENCES term(term_id),      
    FOREIGN KEY (fk_school) REFERENCES school(school_id));''')

df_institute.to_sql('institute', con, index=False, if_exists='append');
df_term.to_sql('term', con, index=False, if_exists='append');
df_school.to_sql('school', con, index=False, if_exists='append');
df_class.to_sql('class', con, index=False, if_exists='append');
```
